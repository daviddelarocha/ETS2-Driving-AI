```python
x_cols = ['truck_speed_kmh', 'speed_limit_kmh', 'cargo_mass_kg']
y_cols = ['steering', 'brake', 'throttle']

df_num = df[x_cols + y_cols].copy()
```

```python
for col in df_num.columns:
    plt.hist(df_num[col])
    plt.title(col)
    plt.show()
```
